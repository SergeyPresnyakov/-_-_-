```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 201)
plt.plot(x, np.cos(1*x), label = 'cos(1*x)')
plt.plot(x, np.cos(2*x),  label = 'cos(2*x)')
plt.xlabel('x')
plt.ylabel('y')


plt.legend(loc = 4)

plt.title('Графики функций cos(1*x) и cos(2*x)', size=12)
plt.show()
```
